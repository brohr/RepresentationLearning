```python

rng = np.random.RandomState(42)

# Generate some 2D coefficients with sine waves with random frequency and phase
n_samples, n_features, n_tasks = 100, 30, 40
n_relevant_features = np.ceil((1/10.) * n_features * n_tasks).astype(int)
coef = np.zeros((n_tasks, n_features)) + np.random.uniform(-0.1,0.1, (n_tasks,n_features)) #/ 10
times = np.linspace(0, 2 * np.pi, n_tasks)
#for k in range(n_relevant_features):
#    coef[:, k] = np.sin((1. + rng.randn(1)) * times + 3 * rng.randn(1))
#    
indi = np.random.randint(0, n_tasks - 1, n_relevant_features)
indj = np.random.randint(0, n_features - 1, n_relevant_features)
for k in range(n_relevant_features):
    coef[indi[k],indj[k]] = np.sin((1. + rng.randn(1)) * times[indi[k]] + 3 * rng.randn(1)) * 2


X = rng.randn(n_samples, n_features)
mask = X > 0
X[~mask] = 0
Y = np.dot(X, coef.T) + rng.randn(n_samples, n_tasks)

coef_lasso_ = np.array([LassoLarsIC().fit(X, y).coef_ for y in Y.T])
coef_multi_task_lasso_ = MultiTaskLasso(alpha=1.).fit(X, Y).coef_
coef_ridge_ = np.array([Ridge(alpha=0.5).fit(X, y).coef_ for y in Y.T])
```

```python

# #############################################################################
# Plot support and time series
fig = plt.figure(figsize=(8, 5))
plt.subplot(1, 4, 1)
plt.imshow(coef_ridge_)
plt.xlabel('Feature')
plt.ylabel('Time (or Task)')
plt.text(10, 5, 'Ridge')
plt.subplot(1, 4, 2)
plt.imshow(coef_lasso_)
plt.xlabel('Feature')
plt.ylabel('Time (or Task)')
plt.text(10, 5, 'Lasso')
plt.subplot(1, 4, 3)
plt.imshow(coef_multi_task_lasso_)
plt.xlabel('Feature')
plt.ylabel('Time (or Task)')
plt.text(10, 5, 'MultiTaskLasso')
plt.subplot(1, 4, 4)
plt.imshow(coef)
plt.xlabel('Feature')
plt.ylabel('Time (or Task)')
plt.text(10, 5, 'Truth')
fig.suptitle('Coefficient non-zero location')

feature_to_plot = 0
plt.figure()
lw = 2
plt.plot(coef[:, feature_to_plot], color='seagreen', linewidth=lw,
         label='Ground truth')
plt.plot(coef_lasso_[:, feature_to_plot], color='cornflowerblue', linewidth=lw,
         label='Lasso')
#plt.plot(coef_multi_task_lasso_[:, feature_to_plot], color='gold', linewidth=lw,
#         label='MultiTaskLasso')
plt.plot(coef_ridge_[:, feature_to_plot], color='maroon', linewidth=lw,
         label='Ridge')
plt.legend(loc='upper center')
plt.axis('tight')
plt.ylim([-2.1, 2.1])
plt.show()
```

```python
print 'Coef Diff Norm: Ridge \t Lasso \n', np.linalg.norm(coef_ridge_ - coef), np.linalg.norm(coef_lasso_ - coef)
```
